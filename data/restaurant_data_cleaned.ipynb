```python
import pandas as pd
```

```python
#Load CSVs

load_restaurant_data = "Independence100.csv"
load_cities_data = "worldcities.csv"
```

```python
#Read CSVs

restaurant_data_df = pd.read_csv(load_restaurant_data)
cities_data_df = pd.read_csv(load_cities_data)
```

```python
#Show data in dataframe

restaurant_data_df
```

```python
#Show data in dataframe

cities_data_df
```

```python
#Drop an duplicate cities

cities_data_df.drop_duplicates(subset=["city", "country"], keep="first", inplace=True)
cities_data_df
```

```python
#Show only cities within the United States

us_cities = cities_data_df.loc[cities_data_df["country"] == "United States"]
us_cities
```

```python
#Remove columns not needed from city data

cities_columns = ["city", "lat", "lng"]
cities_data = us_cities[cities_columns].copy()
cities_data = cities_data.loc[cities_data["city"] != ("Rosemont")]


cities_data
```

```python
#Create dataframe with missing info

nan_cities_df = pd.DataFrame({"city": ["Orlando ", "Ft. Lauderdale", "Bal Harbour", "Cape May", "La Jolla", "Rosemont"],
                              "lat": ["28.4772", "26.1412", "25.8918", "38.9351", "32.8328", "41.9868"],
                              "lng": ["-81.3789", "-80.1373", "-80.1270", "-74.9060", "-117.2713", "-87.8722"]
                             })
nan_cities_df
```

```python
#Append missing info dataframe to cities data

cities_data_append = cities_data.append(nan_cities_df, ignore_index = True)
cities_data_append
```

```python
#Merge restaurant and cities data into one dataframe

restaurant_data_combined = pd.merge(restaurant_data_df, cities_data_append, how="left", left_on="City", right_on="city")
restaurant_data_combined
```

```python
#Rename columns

restaurant_data_cleaned = restaurant_data_combined.rename(columns={"Average Check":"Average_Check", "Meals Served":"Meals_Served", "lat": "Lat", "lng":"Lng"})

restaurant_data_cleaned
```

```python
restaurant_data_cleaned1 = restaurant_data_cleaned.drop(["city"], axis=1)
restaurant_data_cleaned1
```

```python
#Export the dataframe to a CSV

restaurant_data_cleaned1.to_csv("restaurant_data.csv", index=False, header=True)
```

```python
#Export the datafram to HTML
restaurant_data_cleaned1.to_html('data.html', index=False, border=3, justify='center')
```
