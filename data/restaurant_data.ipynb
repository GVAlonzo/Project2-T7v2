```python
import pandas as pd
```

```python
#Load CSVs

load_restaurant_data = "Independence100.csv"
load_cities_data = "worldcities.csv"
```

```python
#Read CSVs

restaurant_data_df = pd.read_csv(load_restaurant_data)
cities_data_df = pd.read_csv(load_cities_data)
```

```python
#Show data in dataframe

restaurant_data_df
```

```python
#Show data in dataframe

cities_data_df.head()
```

```python
#Drop duplicate cities

cities_data_df.drop_duplicates(subset=["city", "country"], keep="first", inplace=True)
len(cities_data_df)
```

```python
#Remove columns not needed from city data

cities_columns = ["city", "lat", "lng"]
cities_data = cities_data_df[cities_columns].copy()

len(cities_data)

```

```python
#Merge 2 CSVs together

restaurant_data = pd.merge(restaurant_data_df, cities_data, how="left", left_on="City", right_on="city")
restaurant_data
    
    
```
